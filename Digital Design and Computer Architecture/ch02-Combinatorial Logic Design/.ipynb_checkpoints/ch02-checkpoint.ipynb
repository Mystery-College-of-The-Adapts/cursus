Introduction
=============

In a digital electronics, **a circuit ** is a network that process discrete valued variables. A circuit can be viewed as a black box of

* **One or more discrete-valued input terminals**
* **One or more discrete-valued output terminals**
* **a functional specification** describing the relationship between inputs and outputs
* **a timing specification** describing the delay between inputs changing and outputs responding.

```python
from IPython.display import Image
Image(filename="assets/box.png")
```

Digital circuits are classified as:
* **Combinatorial** : 
    * outputs depend only on the current values of the inputs
    * is memoryless
    * functional specification expresses the output values in terms of the current input values
    * timing specification consists of lower and upper bounds on the delay from input to output
* **Sequential** : 
    * outputs depend on current and previous values of the inputs
    * has memory
    

Boolean Equations
===================

* **A Product or an Implicant**: the AND of one or more literals
* **A minterm** is a product involving all the inputs to the function.
* **A  maxterm** is a sum involving all the inputs to the function 

**In Boolean equations, NOT has the highest precedence,
followed by AND, then OR.**
