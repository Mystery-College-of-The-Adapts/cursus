Moore & Mealy Machines
==========================

Finite automata may have outputs corresponding to each transition. There are two types of finite state machines that generate
output.:
* **Mealy Machines**
* **Moore Machines**


### Mealy Machine

For a mealy machine, the output depends on the present state as well as the present input( thus, ,the FSM has the output written on edges)


```python
from IPython.display import Image
Image(filename="assets/fsm.png")
```

### Moore Machine
For a moore machine, the output depends on only the present state(thus, the FSM has the output written in the state itself)
