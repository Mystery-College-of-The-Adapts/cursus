# 1.1 Networks and Services
A **Communication network** is a set of equipment and facilities that provides a service- the transfer of information between users located at various geographical points.

Examples of services:
- Radio and television broadcasting
- Telephone service
- Cellular telephone service
- Electronic mail
- Video on demand

# 1.2 Approaches to Network Design
The task of the designer is to develop an overall network design that meets the requirements of the users in a cost-effective manner.

## 1.2.1 Network Functions and Network Topology 
- The essential function of a network is to transfer information between a source and a destination.
- The source and destination typically comprise **terminal** equipment that attaches to the network.
___________________________

- A **Network Switch** serves as a controller, enabling networked devices to talk to each other efficiently.
- **Routing** is the process of selecting a path for traffic in a network, or between or across multiple **networks**.
- **Basic user Service** is the primary service or services that the network provides to its users.
- **Addressing**- the means for identifying points of connection to the network
- **Multiplexing**-the means for connecting multiple information flows into shared connection lines.

## 1.2.2 Message, Packet, and Circuit Switching
- Telephone networks are **connection oriented** because they require the setting up of a connection before the actual transfer of information can take place. 
- **Circuit Switching** is a transfer mode of a network that involves setting up a dedicated end-to-end connection.
- **Packet Switching** is a digital networking communications method that groups all transmitted data into suitably sized blocks; called **packets**, which are transmitted via a medium that may be shared by multiple simultaneous communication sessions.
- ** Message Switching**: messages are routed in their entirety, one hop at a time. Each message is treated as a separate entity.

### Solution:
The steps involved in mailing a letter are:
- The letter is deposited in mailbox.
- The letter is picked up by postal employee and placed in sack.
- The letter is taken to a sorting station, where it is sorted according to destination, as determined by the mail code and grouped with other letters with the same destinationn mail code. (If there is no mail code, then it is determined by the largest geographical unit, for example, country ( if specified), otherwise state (if specified), otherwise city(if specified).
- The letter is shipped to the post office that handles the mail for the specific mail code (or country or city)
- The letter is then sorted by street address
- The letter is picked up at the post office by the postal worker
- The letter is delivered according to the number and street.

### Solution:
The steps involved in e-mailing a message are:
- The message is sent electronically by clicking 'Send'
- The mail provider of the sender sends a request to a name server for the network address of the mail provider of the recipient. The mail provider is determined by the information following the @ symbol.
- If the mail provider finds the network address of the recipient's mail provider, then it sends the message to that address.
- Upon receipt of the message the mail provider either informs the recipient automatically that there is mail or waits until mail is requested by the recipient, depending on how the mail service is set up. The recipient is determined by the information before the @ symbol.
